```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import json
```

```python
listings_df = pd.read_csv('Listing.csv')
listings_df = listings_df.drop(["Unnamed: 0"], axis=1)
listings_df = listings_df.dropna()
```

```python
listings_df = listings_df.loc[listings_df['bodytype'] != 'Other']
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1920x1080")

chrome_driver = "chromedriver.exe"
driver = webdriver.Chrome(chrome_options=chrome_options, executable_path=chrome_driver)

for index, row in listings_df.iterrows():
    
    driver.get("https://www.carqueryapi.com/api/0.3/?callback=?&cmd=getTrims&model=" + str(row['model']) +
              "&year=" + str(row['year']) + '&body=' + str(row['bodytype']))

    results = json.loads(((driver.find_element_by_tag_name('body').text)[2:-2]))
    
    try:
        listings_df.loc[index, "Engine power"] = results['Trims'][0]['model_engine_power_ps']
    except:
        listings_df.loc[index, "Engine power"] = None
    
    try:
        listings_df.loc[index, "Engine type"] = results['Trims'][0]['model_engine_type']
    except:
        listings_df.loc[index, "Engine type"] = None
    
    try:
        listings_df.loc[index, "Engine torque rpm"] = results['Trims'][0]['model_engine_torque_rpm']
    except:
        listings_df.loc[index, "Engine torque rpm"] = None
    
driver.quit()
```

```python
listings_df = listings_df.dropna()
```
